```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
from scipy import ndimage
import imageio.v2 as imageio
```

```python
# 1.  
image = imageio.imread('puppy.jpg')

plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# 2.   
contrast_image = np.clip(image, 50, 200)  #     
plt.imshow(contrast_image)
plt.title('Enhanced Contrast and Resized Image')
plt.axis('off')
plt.show()
```

```python
# 3.       (R, G, B)
colors = ['Reds', 'Greens', 'Blues']
for i, color in enumerate(colors):
    plt.imshow(image[:,:,i], cmap=color)
    plt.title(f'{color[:-1]} Channel')
    plt.axis('off')
    plt.show()
    
```

```python
# 4.    -
gray_image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# 5.  
blurred_image = gaussian_filter(gray_image, sigma=2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```

```python
# 6. Edge detection (  )
sx = ndimage.sobel(blurred_image, axis=0, mode='constant')
sy = ndimage.sobel(blurred_image, axis=1, mode='constant')
edges = np.hypot(sx, sy)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')
plt.show()
```
